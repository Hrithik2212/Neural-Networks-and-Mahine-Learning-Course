# Principal Component Analysis with different Kernels 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

```python
X = df[df.columns[:-1]]
y = df[df.columns[-1]]
```

```python
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)
X_scaled[:5]
```

```python
from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(y)
```

## Linear

```python
from sklearn.decomposition import PCA
PCA_X = PCA(n_components=2).fit_transform(X)
plt.figure(figsize=(8,6))
 
plt.scatter(PCA_X[0:50,0], PCA_X[0:50,1], color='green', marker='o', label='Setosa')
plt.scatter(PCA_X[50:100,0], PCA_X[50:100,1], color='blue', marker='s', label='Versicolour')
plt.scatter(PCA_X[100:150,0], PCA_X[100:150,1], color='red', marker='+', label='Virginica')
 
plt.title("PCA components plot for IRIS Dataset", fontsize=14)
plt.legend()
plt.show()
```

## Radial Basis Function

```python
from sklearn.decomposition import KernelPCA

PCA_X = KernelPCA(n_components=2 , kernel ='rbf').fit_transform(X)
plt.figure(figsize=(8,6))
 
plt.scatter(PCA_X[0:50,0], PCA_X[0:50,1], color='green', marker='o', label='Setosa')
plt.scatter(PCA_X[50:100,0], PCA_X[50:100,1], color='blue', marker='s', label='Versicolour')
plt.scatter(PCA_X[100:150,0], PCA_X[100:150,1], color='red', marker='+', label='Virginica')
 
plt.title("PCA components plot for IRIS Dataset", fontsize=14)
plt.legend()
plt.show()
```

## Polynomial

```python

PCA_X = KernelPCA(n_components=2 , kernel ='poly').fit_transform(X)
plt.figure(figsize=(8,6))
 
plt.scatter(PCA_X[0:50,0], PCA_X[0:50,1], color='green', marker='o', label='Setosa')
plt.scatter(PCA_X[50:100,0], PCA_X[50:100,1], color='blue', marker='s', label='Versicolour')
plt.scatter(PCA_X[100:150,0], PCA_X[100:150,1], color='red', marker='+', label='Virginica')
 
plt.title("PCA components plot for IRIS Dataset", fontsize=14)
plt.legend()
plt.show()
```
